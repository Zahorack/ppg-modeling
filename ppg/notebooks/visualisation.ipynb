```python
import matplotlib.pyplot as plt
```

```python
from ppg.common import config
from ppg.common.utils import load_data
from ppg.enums import DataColumn, WavelengthIndex
```

```python
ts1 = load_data('ppg_curve_0.nc')
ts2 = load_data('ppg_curve_1.nc')
```

```python
ts1[DataColumn.PPG_CURVE][WavelengthIndex.IR]
```

```python
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.plot(ts1[DataColumn.PPG_CURVE][WavelengthIndex.RED])
plt.title('ts1 Red Channel')

plt.subplot(3,1,2)
plt.plot(ts1[DataColumn.PPG_CURVE][WavelengthIndex.IR])
plt.title('ts1 IR Channel')

plt.subplot(3,1,3)
plt.plot(ts1[DataColumn.PPG_CURVE][WavelengthIndex.PLETH])
plt.title('ts1 Pleth Channel')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

START = 1000
END = 1300

plt.subplot(3,1,1)
plt.plot(ts1[DataColumn.TIME][START:END], ts1[DataColumn.PPG_CURVE][WavelengthIndex.RED][START:END], color='red')
plt.title('ts1 Red Channel')
plt.grid()

plt.subplot(3,1,2)
plt.plot(ts1[DataColumn.TIME][START:END], ts1[DataColumn.PPG_CURVE][WavelengthIndex.IR][START:END], color='purple')
plt.title('ts1 IR Channel')
plt.grid()

plt.subplot(3,1,3)
plt.plot(ts1[DataColumn.TIME][START:END], ts1[DataColumn.PPG_CURVE][WavelengthIndex.PLETH][START:END])
plt.title('ts1 Pleth Channel')

plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.plot(ts2[DataColumn.PPG_CURVE][WavelengthIndex.RED])
plt.title('ts2 Red Channel')

plt.subplot(3,1,2)
plt.plot(ts2[DataColumn.PPG_CURVE][WavelengthIndex.IR])
plt.title('ts2 IR Channel')

plt.subplot(3,1,3)
plt.plot(ts2[DataColumn.PPG_CURVE][WavelengthIndex.PLETH])
plt.title('ts2 Pleth Channel')

plt.tight_layout()
plt.show()
```
